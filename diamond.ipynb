```python
import numpy as np
import pandas as pd 
```

```python
train = pd.read_csv('diamonds_train.csv')
train.head()
```

```python
test = pd.read_csv('diamonds_test.csv')
test
```

```python
train['cut'].unique()
```

```python
cut = {'Fair': 0,
           'Very Good' : 1,
           'Good' : 2,
           'Premium' : 3 ,
           'Ideal':4}

train['cut'] = train['cut'].map(cut)
test['cut'] = test['cut'].map(cut)
```

```python
train['color'].unique()
```

```python
train['clarity'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
train['color'] = label_encoder.fit_transform(train['color'])
test['color'] = label_encoder.fit_transform(test['color'])

train['clarity'] = label_encoder.fit_transform(train['clarity'])
test['clarity'] = label_encoder.fit_transform(test['clarity'])
```

```python
train.info()
```

```python
train = train.dropna(axis=0, how='all', subset=['price'])
```

```python
train.columns
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
for i in ['carat', 'cut', 'color', 'clarity', 'depth', 'table', 'x', 'y','z']:
  train[[i]] = scaler.fit_transform(train[[i]])
  test[[i]] = scaler.fit_transform(test[[i]])
```

```python
train
```

```python
y=  train[['price']]
```

```python
x = train.copy()
x.drop(columns= ['price'], inplace= True)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, shuffle=True)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.25, shuffle=True)
```

```python
# modeling
from sklearn.linear_model import LinearRegression

model =LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_val)
```

```python
# evaluate your model on train data 
from sklearn.metrics import r2_score

r2_score(y_val, y_pred)
```

```python
# evaluate model on tast data
y_pred = model.predict(X_test)
r2_score(y_test, y_pred)
```
